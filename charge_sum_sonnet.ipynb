```python
def gr_al(mat,start_x,start_y,queue=[]):
    visited=np.zeros((mat.shape[0],mat.shape[1]), dtype=bool)
    queue.append([start_x,start_y])
    total=0
    while queue :
        i,j = queue.pop()
        ind=0
        #замена mat[i][j] на заряд
        a=mat[i][j].split(' ')[1:3]
        im=float(a[1].split('j')[1])
        real=float(a[0])
        charge0=np.sqrt(im**2+real**2)
        if charge0 != 0:
            if visited[i][j] == False:
                a=mat[i+1][j].split(' ')[1:3]
                im=float(a[1].split('j')[1])
                real=float(a[0])
                charge=np.sqrt(im**2+real**2)
                if charge != 0:
                    queue.append([i+1,j])
                    if ind ==0:
                        visited[i][j]=True
                        ind=1
                        total+=charge0
                a=mat[i][j+1].split(' ')[1:3]
                im=float(a[1].split('j')[1])
                real=float(a[0])
                charge=np.sqrt(im**2+real**2)
                if charge != 0:
                    queue.append([i,j+1])
                    if ind ==0:
                        visited[i][j]=True
                        ind=1
                        total+=charge0
                a=mat[i][j-1].split(' ')[1:3]
                im=float(a[1].split('j')[1])
                real=float(a[0])
                charge=np.sqrt(im**2+real**2)
                if charge != 0:
                    queue.append([i,j-1])
                    if ind ==0:
                        visited[i][j]=True
                        ind=1
                        total+=charge0
                a=mat[i-1][j].split(' ')[1:3]
                im=float(a[1].split('j')[1])
                real=float(a[0])
                charge=np.sqrt(im**2+real**2)
                if charge !=0:
                    queue.append([i-1,j])
                    if ind ==0:
                        visited[i][j]=True
                        ind=1
                        total+=charge0
    print(total)
    return visited
```

```python
vis=gr_al(mat,328,194)
```
